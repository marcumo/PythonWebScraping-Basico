## Web scraping
- Es una técnica utilizada mediante programas de software para extraer información de sitios web.
- Consiste en tres pasos:
![Imagen](images/web-scraping-service.png)


### La extracción de datos
- Desde la web, generalmente desde páginas HTML, correos electrónicos, páginas actualizables con AJAX y archivos ocultos tipo CSV.
- Se utilizan librerías como: urllib(módulo para interactuar con paginas web y hacer peticiones), urllib3, ssl(autenticación de conecciones tipo SSL), imaplib (módulo para ingresar a los correos), email(módulo para interactuar dentro de la bandeja del correo), pandas (para archivos csv)

### Procesamiento de datos
- Se procesan los datos desde los distintos tipos de archivos de tal forma que puedan ser utilizados para extraer información.
- Se utilizan librerías como: re(módulo para hacer las busquedas dentro de textos, buscando palabras clave), BeautifulSoup(módulo para realizar búsquedas en archivos con marcado xlm), Scrapy, Selenium


### Almacenamiento de los datos estructurados
- Los datos esctructurados son almacenados para su posterior utilización ().
- Se Utilizan librerías como: pandas (para almacenamiento de datos en archivos CSV, XLS, XLSX, entre otros), scpy (para almacenar archivos complejos como waves), xlsxwriter, openpyxl, sqlite(interacción con bases de datos SQLite), entre muchos otros

```python
#import sys
#!conda install --yes --prefix {sys.prefix} urllib3

#import urllib3,unicodedata
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import ssl
```

```python
link = 'https://estadisticas.bcrp.gob.pe/estadisticas/series/mensuales/resultados/PN01270PM/html'
```

```python

context = ssl._create_unverified_context()
#urllib.urlopen("https://no-valid-cert", context=context)
html = urlopen(link, context=context).read()#.decode()
 
res = bs(html,"html5lib");
 
print(res.title)
```
